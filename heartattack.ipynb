```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'Heart Attack Data Set.csv'
heart_attack_data = pd.read_csv(file_path)
```

```python
print("Dataset Information:")
heart_attack_data.info()
```

```python
print("\nFirst Few Rows of Data:")
print(heart_attack_data.head())
```

```python
print("\nSummary Statistics:")
print(heart_attack_data.describe())
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = heart_attack_data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(heart_attack_data['age'], kde=True, bins=15, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='chol', data=heart_attack_data, palette="Set2")
plt.title("Cholesterol Levels by Heart Attack Diagnosis")
plt.xlabel("Heart Attack Diagnosis (0 = No, 1 = Yes)")
plt.ylabel("Cholesterol (mg/dL)")
plt.show()
```
